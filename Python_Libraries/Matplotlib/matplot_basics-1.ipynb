```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,30,1000)
y = np.cos(x) + 0.6*x**2
```

```python
plt.xlabel('input')
plt.ylabel('output')
plt.title('Observation')
plt.plot(x,y);
```

## Scatterplot
plots one dot for each observation.

### Example-1

```python
x = ['Zebra', 'Lions', 'Monkeys', 'Elephants', 'Ostriches', 'Snakes', 'Tigers', 'Cheetah']
y = [23, 45, 6, 89, 55, 34, 78, 90]
plt.xlabel('animal type')
plt.ylabel('no. of animals')
plt.title('plot')
plt.scatter(x, y)
plt.show()
```

### Example-2

```python
x = numpy.random.normal(5.0, 1.0, 100)
y = numpy.random.normal(10.0, 2.0, 100)
#print(x,y,sep='\n')
plt.xlabel('input')
plt.ylabel('output')
plt.title('Observation')
plt.scatter(x, y);
```

## Histogram
Also known as `frequency distributions.` It is a graph showing the number of observations within each given interval.

### Example-1

```python
x = numpy.random.normal(5.0, 1.0, 100)
plt.hist(x)
plt.show()
```

## Bars
- bar() - Vertical bars
- barh() - horizontal bars

### Example-1

```python
x = ['Zebra', 'Lions', 'Monkeys', 'Elephants', 'Ostriches', 'Snakes']
y = [23, 45, 6, 89, 55, 34]
plt.xlabel('animal type')
plt.ylabel('no. of animals')
plt.title('plot')
plt.bar(x, y)
plt.show()
```

```python
x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.barh(x, y, color='red')
plt.show()
```
